## Какой роли нет в скраме?

# ИЗ РАЗБОРА

Ответ: Product manager

Почему?
Остальные роли присутствуют в Scrum:

    Scrum master – помогает организовать бизнес-процесс и следит за соблюдением scrum-подхода
    Development team – универсальная команда специалистов
    Product owner – соединяет команду с заказчиком и следит за развитием проекта (= куратор)

## Что не относится к артефактам в Скраме?

# ИЗ РАЗБОРА

Ответ: доска

Почему?

    Продуктовый бэклог – список множества задач и элементов, выполнение которых даст нам готовый продукт.
    Бэклог спринта – это набор элементов бэклога, который команда берет в разработку на ближайший спринт.
    Инкремент – часть проекта или продукта, которая несет собой определенную ценность для заказчика. Т.е. какой-то работающий продукт внутри большого проекта.

## Сопоставьте проекты и подходы к их организации, которые подойдут им лучше всего:



# ИЗ РАЗБОРА

Ответ:  

Небольшой проект с четко описанными требованиями – Waterfall
Объемный проект, у которого нет четкого видения конечного результата – Agile
Почему?: 

    Waterfall – каскадная модель, ключевой особенностью которой является последовательность и ориентация на требования. Она подходит для коротких и несложных проектов, у которых имеются четко установленные требования. 

    Agile лучше всего подходит для проектов, у которых нет чёткой уверенности в том, каким должен быть конечный результат, и которым нужно быстро подстраиваться под изменения. Также нужен для больших, длительных и сложных проектов, с большим количеством идей и неопределенностью.
