Какова вероятность того, что в мишени будет только одна пробоина?

Примечание: в этой и последующих задачах десятичным разделителем должна быть точка. 

```python
Подсказка

- **Что надо иметь в виду при решении?**
    - Определиться, какие ситуации здесь являются взаимоисключающими (зависимыми), а какие нет (независимыми). 
    
    Это однозначно даст вам понять, 
    
    перемножаются ли вероятности или вычитаются/складываются.
    - У задачи четыре исхода: а) не попал никто б) попали оба в) попал только первый стрелок г) попал только второй стрелок. 
        Под вопрос задачи попадают два последних варианта.
```

**Разбор решения**

В данной задаче вам необходимо рассчитать вероятность тех случаев, что один стрелок попадет, а другой стрелок промажет. 
Это можно сделать следующим образом:

1. Рассчитать вероятность того, что оба стрелка попадут в мишень - это будет перемножение вероятностей: $0.5 \times 0.8 = 0.4$
2. Рассчитать вероятность того, что оба стрелка промажут по мишени - это будет перемножение вероятностей промаха: $(1 - 0.5) \times (1 - 0.8) = 0.1$
3. Так как у нас суммарная вероятность всех событей равна 1, то мы можем вычесть вероятность этих двух ситуаций и получить вероятность того, что хотя бы один из стрелков попадет в мишень: $1 - 0.4 - 0.1 = 0.5$

Также можно пойти другим путем и определить вероятности событий:

1. Первый стрелок попал, а второй стрелок промахнулся: $0.5 \times (1 - 0.8) = 0.1$
2. Ситуация наоборот, первый стрелок промахнулся, а второй попал: $(1 - 0.5) \times 0.8 = 0.4$
3. Сложить эти вероятности, так как это две взаимоисключающие ситуации, которые обе складываются в событие "один стрелок попал” $0.1 + 0.4 = 0.5$

##### ОТВЕТ: 0.5 (с 8й попытки) зато сам)

Найдите вероятность того, что хотя бы один аналитик решит задачу.

```python
Подсказка
Ситуация, противоположная "хоть один аналитик решил задачу" - "никто не решил задачу". 
Если сумеете найти вероятность второй ситуации - автоматически найдёте вероятность первой.
```

Решение задачи каждым конкретным аналитиком, событие независимое, поэтому мы можем рассчитать вероятности того, 

решит каждый аналитик задачку или нет:

$q_1 = 1 - p_1 = 1 - 0.8 = 0.2$

$q_2 = 1 - p_2 = 1 - 0.7 = 0.3$

$q_3 = 1 - p_3 = 1 - 0.6 = 0.4$

Тогда по теореме умножения вероятностей независимых событий: $p_{(0)} = q_1 \times q_2 \times q_3 = 0.2 \times 0.3 \times 0.4 = 0.024$ - вероятность того, что будет три промаха. 
    Тогда $1 - p_{(0)} = 1 - 0.024 = 0.976$ – вероятность того, что хотя бы один аналитик сможет решить задачку.

###### ОТВЕТ: 0.976 (уже со второй попытки)

Брошены две игральные кости. Есть следующие события:

А = {выпадение шестерки на первой кости}
В = {сумма выпавших очков равна 7}
С = {выпадение тройки на второй кости}

Какие пары событий являются независимыми? 

Подсказки

- **Что такое "зависимые события"?**
    
    Это когда появление одного события меняет вероятность другого события. 
    
- **А как это можно проверить?**
    
    Пользуясь формулой $P(A \cap B) = P(A)P(B)$, где $\cap$ обозначает символ "И". Человеческим языком: если произведение вероятностей двух событий равно вероятности одновременного наступления этих событий, то они **независимы**.
    
    Попробуйте посчитать вероятность каждого из трёх вариантов, а затем посчитать, с какой вероятностью произойдут два из них одновременно. Есть ли случай, где эти варианты различаются? Или все одинаковы?

###### ОТВЕТ: А и С В и С (7 попытки) тупо перебором(())

##### из разбора

Какие пары событий независимы:

А={выпадение шестерки на первой кости}

В={сумма выпавших очков равна 7}

С={выпадение тройки на второй кости}

Довольно очевидно, что A и C друг от друга не зависят - ведь бросок одной кости никак не меняет результат на другой. Но как быть с A и B, а также B и C? Проверим это через формулу независимости событий:

P(A∩B)=P(A)P(B)

Символ ∩ обозначает здесь "И".

P(A)=1/6 (так как всего граней шесть, а нас интересует всего одно значение)

P(B)=6/36=1/6 (всего комбинаций двух кубиков 36, а условию соответствуют варианты 6 и 1, 5 и 2, 4 и 3, 3 и 4, 2 и 5, 1 и 6)

P(A∩B)=1/36 (если выпала шестерка, то событие B произойдёт только для одной комбинации из 36 - 6 и 1)

Итого: 1/36=1/6∗1/6, что соответствует независимости событий. То же самое справедливо для события C.

Значит, все эти события независимы.

NB! Если бы в качестве события B у нас была любая другая сумма, то тогда A и C/A и B были бы зависимыми. Это связано с тем, что в этом случае количество возможных значений на костях стало бы ограниченным.

Например, сумма, равная двум, может быть выкинута только двумя единицами. Соответственно, наши текущие события A и C исключали бы событие B полностью. Попробуйте также проверить другие варианты - например, 8.
