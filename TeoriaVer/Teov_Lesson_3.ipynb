- **Что надо иметь в виду при решении?**
    
    Здесь нам понадобится лишь частичка формулы Байеса. Вспомните, что:
    
    $$
    P(X) = P(Y_1)P(X|Y_1)+ P(Y_2)P(X|Y_2) +...+P(Y_n)P(X|Y_n)
    $$
    
    P(X) - это как раз вероятность брака компьютера. Остальное связано с вероятностями:
    
    - ...взять **случайный** компьютер из **конкретной** партии (1, 2 или 3)
    - ...получить **бракованный** компьютер, **если** он взят из **конкретной** партии (1, 2 или 3)

$$

Пусть H1 это событие, когда выбирают 1, 2 или третий кубик, а H2 - когда выбирают 4-ый кубик.

Тогда по классическому определению вероятности: 

$$
P(H1) = {3 \over 4}, \ P(H2) = {1 \over 4}
$$

Тогда для события H1 вероятность появления белой грани p = 0.5. 

Посчитаем вероятность того, что при n = 7 (семи бросках) белая грань выпадает ровно k = 1. Используем формулу Бернулли:

$$
P_{7}^{1}=C_{7}^{1}0.5^{1}0.5^{7-1} = 0.0547 = P(A|H1)
$$

Теперь рассматриваем событие H2

Вероятность появления белой грани 

$$
p = \frac{1}{6}
$$

тогда вероятность по формуле Бернулли при n = 7, а k = 1 будет:

$$
P_{7}^{1}=C_{7}^{1}{1 \over 6}^{1}{5 \over 6}^{7-1} = 0.3907 = P(A|H2)
$$

Тогда по формуле полной вероятности получаем:

$$
P(A) = P(H1) \times P(A|H1) + P(H2) \times P(A|H2) = 0.75 \times 0.0547 + 0.25 \times 0.3907 = 0.1387
$$

$$
P(H2|A) = {P(A|H2) \times P(H2) \over P(A)} = {0.25 \times 0.3907 \over 0.1387} = 0.7042
$$

# Задание 6

**Разбор решения**

Пусть,

A - событие, которое соответствует болезни человека ,

H_1 - событие, которое соответствует положительному результату теста

H_2 - событие, которое соответствует отрицательному результату теста

Тогда по формуле Байеса,

$$
P(A|H_1) = \frac{P(H_1|A) * P(A)}{P(H_1)}
$$

$$
P(H_1|A) = 0.99, 
$$

так как мы знаем, что если человек болеет, то тест это покажет с 99 % точностью

$$
P(A) = 0.6 - вероятность \ того, что \ человек \ болен
$$

По формуле полной вероятности,

$$
P(H_1) = P(H_1|A) * P(A) + P(H_1|!A) * P(!A) = 0.99 * 0.6 + 0.01 * 0.4 = 0.598
$$

Подставляем всё в формулу:
$$
P(A|H_1) = \frac{0.99 * 0.6}{0.598} ≈ 0.993
$$
