Подсказки

- **Что надо иметь в виду при решении?**
    
    Здесь нет никаких подвохов - достаточно подставить всё необходимое в формулу Бернулли:
    
    $$
    P_{n}^{k}=C_{n}^{k}p^{k}q^{n-k}
    $$
    
    где 
    
    $$
    C_{n}^{k} = {n! \over k!(n-k)!}
    $$
    
    n в данном случае - количество элементов, k - количество тех элементов, которые должны отказать, p = вероятность отказа, q - вероятность того, что элемент будет работать (1 - p).

```python
round(special.comb(5, 3)*(0.4**3)*(0.6**2),4)
```

##### ОТВЕТ: 0.2304

**Разбор решения**

В данном случае мы можем вспомнить формулу Бернулли:

$$
P_{n}^{k}=C_{n}^{k}p^{k}q^{n-k}
$$

где 

$$
C_{n}^{k} = {n! \over k!(n-k)!}
$$

Тогда мы получаем, что p (вероятность отказа элемента) = 0.4, а число испытаний (т.е. число элементов) n = 5.

Вероятность того, что откажут ровно три элемента из пяти:

$$
P_{5}(3) = C_{5}^3 \times 0.4^3 \times 0.6^2 \approx 0.2304
$$

    Сумма этих вероятностей противоположна вероятности события "попали хотя бы два раза".

```python
# Вероятность общего промаха

(1-0.3)**8
```

**Разбор решения**

В данном случае мы применяем формулу Бернулли:

$$
P_{n}^{k}=C_{n}^{k}p^{k}q^{n-k}
$$

где

$$
C_{n}^{k} = {n! \over k!(n-k)!}
$$

Тогда мы получаем, что p (вероятность попадания в цель) = 0.3, а число испытаний (т.е. число выстрелов) n = 8. Число попаданий в цель в данном случае будет k = 2.

Теперь рассмотрим ситуацию, что мы хотим попасть хотя бы два раза, т.е. нас не устраивают случаи, когда мы совсем не попадаем (k = 0) или попадаем 1 раз (k = 1).

Подставляем значения в формулу Бернулли:



$$
1 - C_{8}^{1} \times 0.3^{1} \times (1-0.3)^{8-1} - C_{8}^{0} \times 0.3^{0} \times (1-0.3)^{8} = 0.745
$$

## Задание 9

В очень большом чате по дата аналитике сидит 1730 человек. Вероятность того, что день рождения наугад взятого участника чата приходится на определенный день года, равна 1/365 для каждого из 365 дней. Найдем наиболее вероятное число счастливчиков, которые родились 1 марта.

Для решения этой задачи вам может потребоваться эта формула.

Подсказки

Биномиальное распределение (распределение по схеме Бернулли) позволяет, в частности, установить, какое число появлений события А наиболее вероятно. Формула для наиболее вероятного числа успехов k (появлений события) имеет вид:

np−q≤k≤np+p,q=1−p.
Так как np−q=np+p−1, то эти границы отличаются на 1. Поэтому k, являющееся целым числом, может принимать либо одно значение, когда np целое число (k=np) , то есть когда np+p (а отсюда и np−q) нецелое число, либо два значения, когда np−q целое число.

- **А это ещё что такое?**
    
    А тут в дело вступает новая формула! Она называется **формулой наивероятнейшего числа успехов** и приложена к заданию по [вот этой ссылке](https://www.matburo.ru/tvart_sub.php?p=calc_bern_naiv).
    
    Сама формула выглядит так:
    
    $$
    np - q \leq m \leq np + p
    $$
    
    В нашем случае $n$ - количество людей в чатике, $p$ - вероятность, что у человека день рождения, $q$ - вероятность, что у человека не день рождения, $m$ - интересующее нас число (должно быть целым).

```python
1730*(1/365) - 364/365
```

```python
1730*(1/365) + 1/365
```

```python
ОТВЕТ: 4
```

**Разбор решения**

В нашем случае,  n = 1730, p = 1/365, q = 364/365

$$
Тогда, np - q \leq k \leq np + p
$$

$$
3.742 \leq k \leq 4.742
$$

k = 4

Ответ: 4
